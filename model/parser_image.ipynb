```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common import NoSuchElementException
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
options.add_argument("--window-size=1920,1080")
```

```python
driver = webdriver.Chrome(options=options)
```

```python
def create_list_image_link(url, count_image=1000, img_selector='img.block', btn_selector='button[data-testid="load-more-button"]') -> list[str]:
    driver = webdriver.Chrome(options=options)
    driver.fullscreen_window()
    driver.get(url) 
    try_find_image = 3
    list_image = []
    
    while True:
        try:
            driver.find_element(By.CSS_SELECTOR, btn_selector)
        except NoSuchElementException:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        else:
            break
        
    while len(list_image) <= count_image or try_find_image == 0:
        try:
            btn = driver.find_element(By.CSS_SELECTOR, btn_selector)
        except NoSuchElementException:
            try_find_image -= 1
        else:
            try:
                btn.click()
            except ElementClickInterceptedException:
                pass
        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        list_image = driver.find_elements(By.CSS_SELECTOR, img_selector)
        print(len(list_image))

    list_image_link = list(map(lambda x: x.get_attribute('src'), list_image))[:count_image]
    driver.quit()
    return list_image_link
```

```python
cat_list_image_link = create_list_image_link('https://ru.freepik.com/search?format=search&last_filter=query&last_value=one+cat&query=one+cat')
```
