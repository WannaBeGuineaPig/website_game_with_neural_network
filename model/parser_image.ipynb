```python
# Импорт библиотек

from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.common.by import By
from selenium.common import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
```

## Создание функции для парсинга сайта freepick

```python
def create_list_image_link(url, count_image=1000, img_selector='img.block', btn_selector='button[data-testid="load-more-button"]') -> list[str]:
    driver = webdriver.Chrome()
    driver.maximize_window()
    driver.get(url) 
    wait = WebDriverWait(driver, 5)
    try_find_image = 3
    list_image = []
    
    while True:
        try:
            driver.find_element(By.CSS_SELECTOR, btn_selector)
        except NoSuchElementException:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        else:
            break
        
    while len(list_image) <= count_image or try_find_image == 0:
        try:
            btn = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, btn_selector)))
        except NoSuchElementException:
            try_find_image -= 1
        except NoSuchElementException:
            pass
        else:
            btn.click()
        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        list_image = driver.find_elements(By.CSS_SELECTOR, img_selector)

    list_image_link = list(map(lambda x: x.get_attribute('src'), list_image))[:count_image]
    driver.quit()
    return list_image_link
```
