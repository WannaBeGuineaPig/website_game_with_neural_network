## Создание функции для парсинга сайта freepick

```python
def create_list_image_link(url, count_image=5000, img_selector='img.object-contain', btn_selector='button[data-testid="load-more-button"]') -> list[str]:
    driver = webdriver.Chrome()
    driver.maximize_window()
    driver.get(url) 
    wait = WebDriverWait(driver, 10)
    try_find_image = 3
    list_image = []
    
    while True:
        try:
            driver.find_element(By.CSS_SELECTOR, btn_selector)
        except NoSuchElementException:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        else:
            break
        
    while len(list_image) <= count_image and try_find_image > 0:
        try:
            btn = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, btn_selector)))
        except NoSuchElementException:
            time.sleep(0.1)
            continue
        try:
            btn.click()

        except ElementClickInterceptedException:
            driver.execute_script("window.scrollTo(0, document.body.Height - 10);")
            continue
        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        if len(list_image) == len(driver.find_elements(By.CSS_SELECTOR, img_selector)):
            try_find_image -= 1
        list_image = driver.find_elements(By.CSS_SELECTOR, img_selector)

    list_image_link = list(map(lambda x: x.get_attribute('src'), list_image))[:count_image]
    driver.quit()
    return list_image_link
```

## Создание функий для сохранения изображений в папку

```python
def save_image_in_folder(link_image: str, path_to_file: str, name_image: str, extension: str = '.png') -> None:
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36 Edg/142.0.0.0',
    }
    response = ''
    attemp = 3
    while response == '':
        if attemp == 0:
            return
        try:
            response = requests.get(link_image, headers=headers)
            break
        except Exception as e:
            time.sleep(0.5)
            attemp -= 1
            print(e)
            
    bytes_image = response.content
    if response.status_code != 200 or len(bytes_image) == 0:
        return
    response.connection.close()
    
    with open(f'{path_to_file}{name_image}{extension}', 'wb') as file:
        file.write(bytes_image)
```

```python
def save_images(list_link_image: list[str], folder_name: str) -> None:
    path = f'./images/{folder_name}/' 
    if not os.path.exists(path):
        os.mkdir(path)
        
    # for i in range(0, len(list_link_image), 2):
    #     for j in range(2):
    #         Thread(target=save_image_in_folder, args=(list_link_image[i + j], path, i + j + 1)).start()
    for i in range(len(list_link_image)):
        save_image_in_folder(list_link_image[i], path, i + 1)
```

```python
def save_all_list(dict_all_list: dict) -> None:
    for dict_data in dict_all_list.values():
        for key, value in dict_data.items():
            Thread(target=save_images, args=(value, key)).start()
```
