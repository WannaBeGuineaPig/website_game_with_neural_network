```python
# Импорт библиотек

from concurrent.futures import ThreadPoolExecutor

from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.common.by import By
from selenium.common import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
```

## Создание функции для парсинга сайта freepick

```python
def create_list_image_link(url, count_image=1000, img_selector='img.block', btn_selector='button[data-testid="load-more-button"]') -> list[str]:
    driver = webdriver.Chrome()
    driver.maximize_window()
    driver.get(url) 
    wait = WebDriverWait(driver, 5)
    try_find_image = 3
    list_image = []
    
    while True:
        try:
            driver.find_element(By.CSS_SELECTOR, btn_selector)
        except NoSuchElementException:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        else:
            break
        
    while len(list_image) <= count_image or try_find_image == 0:
        try:
            btn = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, btn_selector)))
        except NoSuchElementException:
            try_find_image -= 1
        except NoSuchElementException:
            pass
        else:
            btn.click()
        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        list_image = driver.find_elements(By.CSS_SELECTOR, img_selector)

    list_image_link = list(map(lambda x: x.get_attribute('src'), list_image))[:count_image]
    driver.quit()
    return list_image_link
```

## Проверка работоспособности функции

```python
urls = [
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=one+cat&query=one+cat',
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=one+dog&query=one+dog',
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=one+guinea+pig&query=one+guinea+pig',
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=one+cow&query=one+cow',
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=single+chamomile&query=single+chamomile&type=photo',
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=single+rose&query=single+rose&type=photo',
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=single+bell+flower&query=single+bell+flower&type=photo',
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=single+tulip&query=single+tulip&type=photo',
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=single+circle&query=single+circle&type=photo',
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=single+triangle&query=single+triangle&type=photo',
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=single+rectangle&query=single+rectangle&type=photo',
    'https://ru.freepik.com/search?format=search&last_filter=query&last_value=single+square&query=single+square&type=photo'
]
```

```python
list_key = [
    'cat',
    'dog',
    'guinea_pig',
    'cow',
    'chamomile',
    'rose',
    'bell',
    'tilip',
    'circle',
    'triangle',
    'rectangle',
    'square'
]
```

```python
pool = ThreadPoolExecutor()
```

```python
dict_all_list = {}
```

```python
list_data = [pool.submit(create_list_image_link, urls[i]) for i in range(4)]
```

```python
dict_all_list.setdefault('animal', {list_key[i] : list_data[i].result() for i in range(4)})
```

```python
list_data += [pool.submit(create_list_image_link, urls[i]) for i in range(4, 8)]
```

```python
dict_all_list.setdefault('plant', {list_key[i] : list_data[i].result() for i in range(4, 8)})
```

```python
list_data += [pool.submit(create_list_image_link, urls[i]) for i in range(8, 12)]
```

```python
dict_all_list.setdefault('figure', {list_key[i] : list_data[i].result() for i in range(8, 12)})
```
